```python
claim=pd.read_csv('Claim.csv')
Demographics=pd.read_csv('Demographics.csv')
Fraud_data=pd.read_csv('Fraud Data.csv')
Policy=pd.read_csv('Policy.csv')
Vehicle=pd.read_csv('Vehicle.csv')
```

```python
newdf1 = Vehicle[(Vehicle['VehicleAttribute']=='VehicleID')].drop(['VehicleAttribute'],axis=1).rename(columns = {'VehicleAttributeDetails':'VehicleID'})
newdf2 = Vehicle[(Vehicle['VehicleAttribute']=='VehicleModel')].drop(['VehicleAttribute'],axis=1).rename(columns = {'VehicleAttributeDetails':'VehicleModel'})
newdf3 = Vehicle[(Vehicle['VehicleAttribute']=='VehicleMake')].drop(['VehicleAttribute'],axis=1).rename(columns = {'VehicleAttributeDetails':'VehicleMake'})
newdf4 = Vehicle[(Vehicle['VehicleAttribute']=='VehicleYOM')].drop(['VehicleAttribute'],axis=1).rename(columns = {'VehicleAttributeDetails':'VehicleYOM'})
```

```python
Vehicledata=newdf1.merge(newdf2,how='outer',on='CustomerID').merge(newdf3,how='outer',on='CustomerID').merge(newdf4,how='outer',on='CustomerID')
```

```python
Vehicledata
```

```python
datamerge=Demographics.merge(Policy,how='outer',on='CustomerID').merge(claim,how='outer',on='CustomerID').merge(Vehicledata,how='outer',on='CustomerID').merge(Fraud_data,how='outer',on='CustomerID')
```

```python
missing_values=['?',-5,'MISSINGVALUE','MISSEDDATA','NA',-1,'???']
datamerge=datamerge.replace(missing_values,np.NaN)
datamerge
```

```python
datamerge=datamerge.drop (['Country','CustomerID','DateOfPolicyCoverage','DateOfIncident','IncidentAddress','VehicleID'],axis=1)
datamerge
```

```python
datamerge['InsuredGender'] = datamerge['InsuredGender'].fillna(datamerge['InsuredGender'].mode()[0])
datamerge['PolicyAnnualPremium'] = datamerge['PolicyAnnualPremium'].fillna(datamerge['PolicyAnnualPremium'].mode()[0])
datamerge['TypeOfCollission'] = datamerge['TypeOfCollission'].fillna(datamerge['TypeOfCollission'].mode()[0])
datamerge['IncidentTime'] = datamerge['IncidentTime'].fillna(datamerge['IncidentTime'].median())
datamerge['PropertyDamage'] = datamerge['PropertyDamage'].fillna(datamerge['PropertyDamage'].mode()[0])
datamerge['Witnesses'] = datamerge['Witnesses'].fillna(datamerge['Witnesses'].median())
datamerge['PoliceReport'] = datamerge['PoliceReport'].fillna(datamerge['PoliceReport'].mode()[0])
datamerge['AmountOfTotalClaim'] = datamerge['AmountOfTotalClaim'].fillna(datamerge['AmountOfTotalClaim'].median())
datamerge['VehicleMake'] = datamerge['VehicleMake'].fillna(datamerge['VehicleMake'].mode()[0])
```

```python
datamerge
```

```python
datamerge.isna().sum()
```

```python
datamerge['VehicleYOM'] = pd.to_numeric(datamerge['VehicleYOM'])
datamerge['AmountOfTotalClaim'] = pd.to_numeric(datamerge['AmountOfTotalClaim'])
datamerge['Witnesses'] = pd.to_numeric(datamerge['Witnesses'])
datamerge['ReportedFraud'] = datamerge['ReportedFraud'].map({'Y': 1, 'N': 0})
```

```python
datamerge['ReportedFraud'].value_counts()
```

```python
dummies=['InsuredGender','InsuredEducationLevel','InsuredOccupation','InsuredHobbies','InsurancePolicyState','Policy_CombinedSingleLimit','InsuredRelationship','TypeOfIncident','TypeOfCollission','SeverityOfIncident','AuthoritiesContacted','IncidentState','IncidentCity','PropertyDamage','PoliceReport','VehicleModel','VehicleMake']
```

```python
final_data = pd.get_dummies(datamerge,columns=dummies,drop_first=True)
final_data
```

```python
final_data.info()
```

```python
x=final_data.drop(['ReportedFraud'],axis=1)
y=final_data['ReportedFraud']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
predictions=lm.predict(x_test)
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,predictions))
print('MSE:',metrics.mean_squared_error(y_test,predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
print('R2:',metrics.r2_score(y_test,predictions))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

# DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(x_train,y_train)
```

```python
predictions=dtree.predict(x_test)
```

```python
print(classification_report(y_test,predictions))
```

# RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=500)
```

```python
rfc.fit(x_train,y_train)
```

```python
predictions=rfc.predict(x_test)
print(classification_report(y_test,predictions))
```

# GaussianNB
